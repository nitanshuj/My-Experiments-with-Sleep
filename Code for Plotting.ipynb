```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("G:\My Drive\& Semester 3 - Fall 2022\INFO-I507-Health Informatics\Homework 1\Data\SLEEP\SLEEP_1664639678208.csv")
df.head()
```

```python
df.shape
```

```python
df['deepSleepHours'] = df['deepSleepTime']/60
df['shallowSleepHours'] = df['shallowSleepTime']/60
df['totalTimeInBed'] = df['deepSleepTime'] + df['shallowSleepTime']
df['totalTimeInBedHours'] = df['totalTimeInBed']/60
#df.head()
```

```python
df.head()
```

```python
# Days my sleep was not recorded
df_days_Unrecorded = df[df['totalTimeInBedHours']==0] 
print("Number of Days NOT recorded ==> ",df_days_Unrecorded.shape[0])

# Days my sleep was recorded
df_days_Recorded = df[df['totalTimeInBedHours']!=0] 
print("Number of Days recorded     ==> ",df_days_Recorded.shape[0])
```

```python
# Number of Deepsleep Hours Slept vs Dates
# ----------------------------------------
plt.figure(figsize=(15,10))
sns.set(style="darkgrid", palette="muted", color_codes=True)
sns.lineplot(x=df_days_Recorded['date'], y=df_days_Recorded['deepSleepHours'])
plt.xlabel("Dates")
plt.ylabel("Number of Deepsleep Hours")
plt.title("Number of Deepsleep Hours Slept vs Dates", size=20)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Number of Deepsleep Hours Slept vs Dates
# ----------------------------------------
plt.figure(figsize=(15,10))
sns.set(style="darkgrid", palette="muted", color_codes=True)
#sns.lineplot(x=df_days_Recorded['date'], y=df_days_Recorded['deepSleepHours'])
sns.lineplot(x=df_days_Recorded['date'], y=df_days_Recorded['shallowSleepHours'], )
plt.xlabel("Dates")
plt.ylabel("Number of Shalow Sleep Hours")
plt.title("Number of Shallow Sleep Hours vs Dates", size=20)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Number of Hours Slept vs Dates
# ------------------------------
plt.figure(figsize=(15,10))
sns.set(style="darkgrid", palette="muted", color_codes=True)
sns.lineplot(x=df_days_Recorded['date'], y=df_days_Recorded['totalTimeInBedHours'], label = 'Total Sleeping Time')
sns.lineplot(x=df_days_Recorded['date'], y=df_days_Recorded['deepSleepHours'], label = 'Deep Sleep Time')
sns.lineplot(x=df_days_Recorded['date'], y=df_days_Recorded['shallowSleepHours'], label = 'Light/Shallow Sleep Time')
plt.xlabel("Dates")
plt.ylabel("Number of Hours")
plt.title("Number of Hours Slept vs Dates", size=20)
plt.xticks(rotation = 90)
plt.show()
```
